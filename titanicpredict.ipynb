```python
df.columns
df.columns=['PID', 'Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch',
       'Ticket', 'Fare', 'Cabin', 'Embarked']
```

```python
df.head(30)
```

```python
int(df['Age'].mean())
```

```python
df.Age.fillna(int(df['Age'].mean()), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.Embarked.mode()[0]
```

```python
df.Embarked.fillna(df.Embarked.mode()[0], inplace=True)
```

```python
df.Cabin.head(30)
```

```python
df.Cabin.mode()
```

```python
df.Cabin.fillna(df.Cabin.mode()[2], inplace=True)
```

```python
df.head(30)
```

```python
def f(s):
    if s=='male':
        return 0
    else:
        return 1
    
df["Gender"] = df.Sex.apply(f)
df.drop('Sex',axis=1, inplace=True)
```

```python
fig= plt.figure(figsize=(10,5))
plt.scatter(df.Survived, df.Age)
plt.xlabel('Survival Rate')
plt.ylabel('Age')
plt.title('Age wrt Survived')
plt.show()
```

```python
df.Survived.value_counts()
fig= plt.figure(figsize=(5,5))
df.Survived.value_counts().plot(kind='bar', alpha=1,color=['green','red'], grid=False)
plt.xlabel('0 - Survived, 1 - Not Survived')
plt.ylabel('No. of peoples')
plt.show()
```

```python
df.Pclass.value_counts()
```

```python
df.Pclass.value_counts()
fig= plt.figure(figsize=(10,5))
df.Pclass.value_counts().plot(kind='bar',alpha=1,color=['red','blue','green'])
plt.show()
```

```python
#correlation between two different features.
#one is scatter plot which we seen above and another one is KDE (Kernal Desnsity estimation)

for i in [1,2,3]:
    df.Age[df.Pclass==i].plot(kind='kde', linewidth=3, alpha=1, grid=True)
    
plt.legend(('1st'))
```
